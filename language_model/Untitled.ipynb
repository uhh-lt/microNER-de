```python
from collections import defaultdict, Counter
import re
```

```python
temp = defaultdict(int)
temp2=[]
temp3=[]
off = defaultdict(int)
other =defaultdict(int)
```

```python
with open("/home/gwiedemann/notebooks/OffLang/sample_train.txt") as f:
    for line in f:
        word , label1 , label2 = line.split('\t')
        user = word.split()[0]
        if re.match('@[\\w_]+', user):
            if label1 == 'OFFENSE':
                off[user]+=1
            else:
                other[user]+=1

        
```

```python
prob=[]
user_set =set()
for x in [off.keys(), other.keys()]:
    for user in x:
        user_set.add(user)
print(len(user_set))
check=set()
for user in user_set:
        check.add(user)
        prob_off = off[user]/(off[user] + other[user])
        prob.append((user,off[user], other[user], prob_off))
#     print((user, off[user], other[user], prob_off))
print(len(check))
```
